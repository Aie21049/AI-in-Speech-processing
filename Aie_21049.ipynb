```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

filename = r"C:\Users\Downloads\WhatsApp Audio 2024-02-08 at 14.10.18.wav"
y, sr = librosa.load(filename)

plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr, color='b') 

plt.title('Speech Signal Waveform')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.show() 
```

```python
length_seconds = librosa.get_duration(y=y, sr=sr)
magnitude_range = max(y) - min(y)
print(f"Length of the signal: {length_seconds:.2f} seconds")
print(f"Magnitude range of the signal: {magnitude_range:.2f}")
```

```python
segment_start = 1
segment_duration = 5.45  
segment = y[int(segment_start * sr):int((segment_start + segment_duration) * sr)]
ipd.display(ipd.Audio(segment, rate=sr)) 
```

```python
start_time = 1
end_time = 4.45
y_abrupt = y.copy()
y_abrupt[int(start_time * sr):int(end_time * sr)] = 0
 
plt.figure(figsize=(10, 4))
plt.plot(y, label='Original')
plt.plot(y_abrupt, label='Abruptly Segmented')
plt.title('Comparison of Original and Abruptly Segmented Speech Signals')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.legend()
plt.show() 
```
